
1: Unique Words in a Sentence


```python
# Take input from user
sentence = input("Enter sentence: ")

# Convert sentence into list of words
words = sentence.split()

# Convert list into set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2: Highest Salary from Employee Data

```python
# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Print result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

3: Find Maximum and Minimum Values


```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

 4: Count Products Above a Price Threshold



```python
# Given list
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

5: Calculate Attendance Percentage


```python
# Given list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

6: Remove Duplicate Phone Numbers

```python
# Given list
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

7: Count Character Frequency


```python
# Given string
text = "python"

# Create empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

# Print result
print(char_count)
```

 9: Check if a Key Exists in a Dictionary


```python
# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
